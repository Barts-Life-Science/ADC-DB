```python
%sql
SELECT EVENT_TITLE_TEXT, EVENT_CD, MAX(cd.DESCRIPTION), COUNT(DISTINCT CLINICAL_EVENT_ID), COUNT(DISTINCT EVENT_ID)
FROM 4_prod.raw.mill_clinical_event AS ce 
INNER JOIN 4_prod.raw.mill_encounter AS en ON ce.encntr_id = en.encntr_id 
LEFT JOIN 3_lookup.mill.mill_code_value AS cd ON ce.EVENT_CD = cd.CODE_VALUE
WHERE 
EVENT_TITLE_TEXT ILIKE '%IV sedation%' 
AND MED_SERVICE_CD = 3768938 
AND COALESCE(EVENT_START_DT_TM, EVENT_END_DT_TM) BETWEEN '2024-09-01' AND '2025-09-30'
GROUP BY EVENT_TITLE_TEXT, EVENT_CD
ORDER BY COUNT(DISTINCT CLINICAL_EVENT_ID) DESC
LIMIT 10
```
