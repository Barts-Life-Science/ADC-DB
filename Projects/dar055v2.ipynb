```python
%sql

CREATE SCHEMA IF NOT EXISTS 5_projects.dar055
```

```python
%sql
CREATE OR REPLACE TABLE 5_projects.dar055.rde_emergencyd AS
SELECT * EXCEPT (MRN, NHS_NUMBER)
FROM 4_prod.rde.rde_emergencyd
WHERE (
    Arrival_Dt_Tm BETWEEN '2018-01-01' AND '2025-12-01'
    OR Departure_Dt_Tm BETWEEN '2018-01-01' AND '2025-12-01'
)
```

```python
%sql
CREATE OR REPLACE TABLE 5_projects.dar055.rde_patient_demographics AS
SELECT t.* EXCEPT (MRN, NHS_NUMBER, Date_of_birth)
FROM 4_prod.rde.rde_patient_demographics AS t
INNER JOIN (SELECT DISTINCT PERSON_ID FROM 5_projects.dar055.rde_emergencyd) AS s
ON t.PERSON_ID = s.PERSON_ID
```

```python
%sql
CREATE OR REPLACE TABLE 5_projects.dar055.rde_all_problems AS
SELECT t.* EXCEPT (MRN, NHS_NUMBER)
FROM 4_prod.rde.rde_all_problems AS t
INNER JOIN (SELECT DISTINCT PERSON_ID FROM 5_projects.dar055.rde_emergencyd) AS s
ON t.PERSON_ID = s.PERSON_ID
```

```python
%sql
CREATE OR REPLACE TABLE 5_projects.dar055.rde_all_diagnosis AS
SELECT t.* EXCEPT (MRN, NHS_NUMBER)
FROM 4_prod.rde.rde_all_diagnosis AS t
INNER JOIN (SELECT DISTINCT PERSON_ID FROM 5_projects.dar055.rde_emergencyd) AS s
ON t.PERSON_ID = s.PERSON_ID
```
