```python
%sql
SELECT Diag_SNOMED_Desc, COUNT(*), COUNT(DISTINCT PERSON_ID)
FROM 4_prod.rde.rde_emergencyd 
WHERE
  (
    Diag_SNOMED_Desc ILIKE 'dental%' 
    OR DIag_SNOMED_Desc ILIKE '% dental%'
    OR Diag_SNOMED_Desc ILIKE '%tooth%'
    OR Diag_SNOMED_Desc ILIKE '%teeth%'
  )
  AND 
  (
    Arrival_Dt_Tm BETWEEN '2018-01-01' AND '2025-12-01'
    OR Departure_Dt_Tm BETWEEN '2018-01-01' AND '2025-12-01'
  )
GROUP BY Diag_SNOMED_Desc
LIMIT 100
```

```python
%sql
CREATE OR REPLACE TABLE 5_projects.dar055.rde_emergencyd AS
SELECT * EXCEPT (MRN, NHS_NUMBER)
FROM 4_prod.rde.rde_emergencyd
WHERE (
    Diag_SNOMED_Desc ILIKE 'dental%' 
    OR Diag_SNOMED_Desc ILIKE '% dental%'
    OR Diag_SNOMED_Desc ILIKE '%tooth%'
    OR Diag_SNOMED_Desc ILIKE '%teeth%'
)
AND (
    Arrival_Dt_Tm BETWEEN '2018-01-01' AND '2025-12-01'
    OR Departure_Dt_Tm BETWEEN '2018-01-01' AND '2025-12-01'
)
```
