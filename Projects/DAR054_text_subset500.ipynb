```python
%sql
CREATE TABLE 5_projects.dar054.rde_blobdataset_subset500 AS
SELECT * FROM 5_projects.dar054.rde_blobdataset LIMIT 500
```

```python
import os
```

```python

base_path = "/Volumes/5_projects/dar054/text_data"  # Or abfss:// for ADLS

# Function to write each row
def write_row(row):
    file_path = os.path.join(base_path, f"eventid={row['eventid']}.txt")
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(row["anonymizedtext"])

df = spark.table("5_projects.dar054.rde_blobdataset_subset500")
# Apply function to each row without collecting
df.select("eventid", "anonymizedtext").foreach(write_row)
```
