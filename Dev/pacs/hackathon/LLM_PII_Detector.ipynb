```python
from openai import OpenAI
import base64
import sys
import pydicom
from PIL import Image
import io
import numpy as np
import json
import os
from tqdm.notebook import tqdm

endpoint = "https://telefonica-hackathon-2026.cognitiveservices.azure.com/openai/v1/"
model_name = "gpt-4.1-mini"
deployment_name = "gpt-4.1-mini"





def dicom_to_base64_png(dicom_path: str) -> str:
    """
    Convert a DICOM file to a PNG image in memory and return Base64 string.

    Args:
        dicom_path (str): Path to the DICOM file.

    Returns:
        str: Base64-encoded PNG image suitable for LLM input.
    """

    # Load DICOM
    ds = pydicom.dcmread(dicom_path)

    # Extract pixel array
    pixel_array = ds.pixel_array.astype(np.float32)

    # Normalize to 0-255
    pixel_array -= pixel_array.min()
    pixel_array /= pixel_array.max()
    pixel_array *= 255.0
    pixel_array = pixel_array.astype(np.uint8)

    # Convert to PIL image
    image = Image.fromarray(pixel_array)

    # Save PNG to memory buffer
    buffer = io.BytesIO()
    image.save(buffer, format="PNG")

    # Convert to Base64
    base64_str = base64.b64encode(buffer.getvalue()).decode("utf-8")

    return base64_str



def base64_to_llm(base64_img, system_prompt, user_prompt):
  client = OpenAI(
              base_url=f"{endpoint}",
              api_key=api_key
          )

  completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
          "role": "system",
          "content": system_prompt,
        },
        {
        "role": "user",
        "content": [
                  {"type": "text", "text": user_prompt},

                  {

                      "type": "image_url",

                      "image_url": {

                          "url": f"data:image/png;base64,{base64_img}"

                      }

                  }

              ]
        }
  ,
      ],
    )
  return completion

def ask_llm(png_file, system_prompt, user_prompt):
    completion = base64_to_llm(png_to_base64(png_file), system_prompt, user_prompt)
    return completion



```

```python
def proc_func(dicom_file):
    completion = base64_to_llm(dicom_to_base64_png(dicom_file), system_prompt, user_prompt)
    result = json.loads(completion.choices[0].message.content)
    result["filepath"] = dicom_file
    return result
```

```python
def iterate_thru_files(root_dir, output_txt):
    """
    Recursively search for PNG files and call proc_func on each.
    """
    results = []

    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in tqdm(filenames):
            tqdm.write(f"processing: {filename}")
            if filename.lower().endswith(".dcm") and not "DICOMDIR" in filename:
                full_path = os.path.join(dirpath, filename)
                try:
                    r = proc_func(full_path)
                    results.append(r)
                except:
                    pass
                #with open(output_txt, 'w') as f:
                #    f.write(str(r))

    return results
```
