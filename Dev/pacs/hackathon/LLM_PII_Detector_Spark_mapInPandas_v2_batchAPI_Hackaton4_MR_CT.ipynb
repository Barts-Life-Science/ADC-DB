```python
import time
import pandas as pd
from pathlib import Path
from pydicom.encaps import get_frame, generate_pixel_data_frame
import gc
import uuid
import json

api_key = dbutils.secrets.get(scope = "adc_store", key = "image-anon_gpt41minibatch_api-key")
endpoint = dbutils.secrets.get(scope = "adc_store", key = "image-anon_gpt41minibatch_endpoint-url")


def extract_dicom_tags(file_path):
    ds_header = pydicom.dcmread(file_path, stop_before_pixels=True)
    accession_number = ds_header.get("AccessionNumber", None)
    modality = ds_header.get("Modality", None)
    return accession_number, modality


def extract_first_frame(file_path):
    """Extract frame 0 from any DICOM. Memory-safe for multi-frame files."""
    # Read header to check frame count
    ds_header = pydicom.dcmread(file_path, stop_before_pixels=True)
    rows = ds_header.Rows
    cols = ds_header.Columns
    bits = ds_header.BitsAllocated
    num_frames = int(ds_header.NumberOfFrames) if "NumberOfFrames" in ds_header else 1
    samples = ds_header.SamplesPerPixel if "SamplesPerPixel" in ds_header else 1

    del ds_header

    # Estimate full uncompressed size
    full_size_mb = rows * cols * (bits // 8) * num_frames * samples / (1024**2)

    ds = pydicom.dcmread(file_path, stop_before_pixels=False)

    if num_frames > 1 and full_size_mb > 50:
        # Large multi-frame: extract only first compressed frame
        is_encapsulated = (
            hasattr(ds["PixelData"], "is_undefined_length")
            and ds["PixelData"].is_undefined_length
        )
        if is_encapsulated:
            try:
                frame_bytes = get_frame(ds.PixelData, 0)
            except Exception:
                frame_bytes = next(generate_pixel_data_frame(ds.PixelData, num_frames))
            del ds
            gc.collect()
            return np.array(Image.open(BytesIO(frame_bytes)))
        else:
            # Uncompressed multi-frame: slice raw bytes for frame 0
            frame_nbytes = rows * cols * (bits // 8) * samples
            raw = ds.PixelData[:frame_nbytes]
            del ds
            dtype = np.uint8 if bits == 8 else np.uint16
            return np.frombuffer(raw, dtype=dtype).reshape(rows, cols)
    else:
        # Single frame or small multi-frame: safe to decompress all
        arr = ds.pixel_array
        if arr.ndim == 3 and num_frames > 1:
            frame = arr[0].copy()
            del arr
        else:
            frame = arr
        del ds
        return frame
    
def frame_to_b64(frame_arr):
    """Resize frame to 512x512 and encode as base64 PNG."""
    img = Image.fromarray(frame_arr)
    del frame_arr
    img = img.resize((512, 512), Image.LANCZOS)

    arr_small = np.array(img, dtype=np.float32)
    del img
    arr_min, arr_max = arr_small.min(), arr_small.max()
    if arr_max > arr_min:
        arr_small = ((arr_small - arr_min) * (255.0 / (arr_max - arr_min))).astype(np.uint8)
    else:
        arr_small = np.zeros(arr_small.shape, dtype=np.uint8)

    img_out = Image.fromarray(arr_small)
    del arr_small
    buf = BytesIO()
    img_out.save(buf, format="PNG", optimize=True)
    b64 = base64.b64encode(buf.getvalue()).decode("utf-8")
    buf.close()
    del img_out
    return b64

def generate_json_line(custom_id, b64):
    deployment_name = "gpt-4.1-mini-batch"

    system_prompt = "Output the result in a structured format with 4 required fields: is_medical_scan, is_text_detected, is_personal_information_detected (e.g. name, ID numbers), and confidence_score. Valid values for is_medical_scan are yes, no, and not_sure. Answer yes only if the input image is a typical medical scan of an anatomical structure. Answer no if the input image is a screen capture, graph, or document etc. Valid values for is_text_detected are yes, no, and not_sure. Valid values for is_personal_information_detected are yes, no, and not_sure. confidence_score is an integer from 0 to 100. 0 indicates high confidence of no text detected. 100 indicates high confidence of text being detected. 50 indicates uncertain. Do not leave any field empty. Always respond with 4 fields everytime."

    user_prompt = "Tell me if there is any text in the image. Also tell me if these texts contain any personal information e.g. name, date of birth, scan date, ID numbers. Then give me a confidence score."

    response_format = {
        "type": "json_schema",
        "json_schema": {
            "name": "PIIDetectionResult",
            "strict": True,
            "schema": {
                "type": "object",
                "properties": {
                    "is_text_detected": {"type": "boolean"},
                    "is_personal_information_detected": {"type": "boolean"},
                    "confidence_score": {"type": "number"},
                    "detected_text_summary": {"type": "string"},
                },
                "required": [
                    "is_text_detected",
                    "is_personal_information_detected",
                    "confidence_score",
                    "detected_text_summary",
                ],
                "additionalProperties": False,
            },
        },
    }

    line = json.dumps({
        "custom_id": custom_id,
        "method": "POST",
        "url": "/v1/chat/completions",
        "body": {
            "model": deployment_name,
            "messages": [
                {"role": "system", "content": system_prompt},
                {
                    "role": "user",
                    "content": [
                        {"type": "text", "text": user_prompt},
                        {
                            "type": "image_url",
                            "image_url": {
                                "url": f"data:image/png;base64,{b64}",
                                "detail": "low",
                            },
                        },
                    ],
                },
            ],
            "max_tokens": 200,
            "response_format": response_format,
        },
    }) + "\n"
    return line

def llm_pii_detection_upload_batch(pdf_iter):





    client = OpenAI(base_url=endpoint,api_key=api_key)

    for pdf in pdf_iter:
        guid = str(uuid.uuid4())
        redacted_rows = []
        #for path, content in zip(pdf['path'], pdf['content']):
        json_lines = ""
        for path in pdf['path']:
            try:
                start_time_str = time.strftime("%Y-%m-%d %H:%M:%S")
                is_loaded_ok = False
                accession_number = None
                modality = None
                

                path = path[5:]
                task_id = path.replace(input_dir, "")

                frame = extract_first_frame(path)
                base64_str = frame_to_b64(frame)
                del frame
 
                is_loaded_ok = True

                line = generate_json_line(task_id, base64_str)
                json_lines += line

                accession_number, modality = extract_dicom_tags(path)
                
                redacted_rows.append((path, is_loaded_ok, accession_number, modality, None, start_time_str, time.strftime("%Y-%m-%d %H:%M:%S"), guid))

            except Exception as e:
                redacted_rows.append((path, is_loaded_ok, accession_number, modality, str(e)[:1000], start_time_str, time.strftime("%Y-%m-%d %H:%M:%S"), guid))

        try:
            #jsonl_filename = task_id+".jsonl"
            file_response = client.files.create(
                file=(guid+".jsonl", json_lines.encode("utf-8")),
                purpose="batch",
                extra_body={"expires_after": {"seconds": 1209600, "anchor": "created_at"}},
            )
            file_id = file_response.id

            batch_response = client.batches.create(
                input_file_id=file_id,
                endpoint="/chat/completions",
                completion_window="24h",
            )

            redacted_rows = [row+(file_id, batch_response.id, batch_response.status) for row in redacted_rows]
        except Exception as e:
            redacted_rows = [row+(None, None, str(e)[:1000]) for row in redacted_rows]




        yield pd.DataFrame(redacted_rows, columns=["path", "is_loaded_ok", "accession_number", "modality", "error_message", "start_ts", "end_ts", "batch_guid", "file_id", "batch_response_id", "batch_response_status"])
```

```python
def parse_output_line(line):
    resp = json.loads(line)
    custom_id = resp["custom_id"]

    if resp.get("error"):
        return {
            "custom_id": custom_id,
            "is_text_detected": None,
            "is_pii_detected": None,
            "confidence_score": None,
            "detected_text_summary": None,
            "llm_raw_result": None,
            "error_message": json.dumps(resp["error"]),
        }

    try:
        raw_content = resp["response"]["body"]["choices"][0]["message"]["content"]
        parsed = json.loads(raw_content)
        return {
            "custom_id": custom_id,
            "is_text_detected": parsed["is_text_detected"],
            "is_pii_detected": parsed["is_personal_information_detected"],
            "confidence_score": float(parsed["confidence_score"]),
            "detected_text_summary": parsed["detected_text_summary"],
            "llm_raw_result": raw_content,
            "error_message": None,
        }
    except Exception as e:
        return {
            "custom_id": custom_id,
            "is_text_detected": None,
            "is_pii_detected": None,
            "confidence_score": None,
            "detected_text_summary": None,
            "llm_raw_result": str(resp.get("response", "")),
            "error_message": f"Parse error: {str(e)[:500]}",
        }
```

```python
from pyspark.sql.functions import col, explode, split, when, udf
from pyspark.sql.types import StringType, BooleanType, FloatType, StructType, StructField

def parse_output_content(output_content):
    import json
    results = []
    if output_content is None:
        return results
    for line in output_content.splitlines():
        try:
            resp = json.loads(line)
            custom_id = resp.get("custom_id")
            if resp.get("error"):
                results.append((custom_id, None, None, None, None, None, json.dumps(resp["error"])))
                continue
            raw_content = resp["response"]["body"]["choices"][0]["message"]["content"]
            parsed = json.loads(raw_content)
            results.append((
                custom_id,
                parsed.get("is_text_detected"),
                parsed.get("is_personal_information_detected"),
                float(parsed.get("confidence_score")) if parsed.get("confidence_score") is not None else None,
                parsed.get("detected_text_summary"),
                raw_content,
                None
            ))
        except Exception as e:
            results.append((custom_id if 'custom_id' in locals() else None, None, None, None, None, None, f"Parse error: {str(e)[:500]}"))
    return results

parse_schema = StructType([
    StructField("custom_id", StringType(), True),
    StructField("is_text_detected", BooleanType(), True),
    StructField("is_pii_detected", BooleanType(), True),
    StructField("confidence_score", FloatType(), True),
    StructField("detected_text_summary", StringType(), True),
    StructField("llm_raw_result", StringType(), True),
    StructField("error_message", StringType(), True)
])

parse_output_content_udf = udf(parse_output_content, "array<struct<custom_id:string,is_text_detected:boolean,is_pii_detected:boolean,confidence_score:double,detected_text_summary:string,llm_raw_result:string,error_message:string>>")

```

```python
MAX_BATCH = 5
for batch_idx in tqdm(range(MAX_BATCH), desc="Batches"):
    df_retrieve_path = f"/Volumes/1_inland/sectra/vone/{str(dbutils.widgets.get('output_parquet'))}_retrieve_batch{batch_idx}.parquet"
    df_retrieve = spark.read.parquet(df_retrieve_path)

    df_parsed = (
        df_retrieve
        .filter(col("error_message").isNull() & col("output_content").isNotNull())
        .withColumn("parsed", parse_output_content_udf(col("output_content")))
        .withColumn("parsed", explode("parsed"))
        .select(
            col("batch_response_id"),
            col("parsed.custom_id"),
            col("parsed.is_text_detected"),
            col("parsed.is_pii_detected"),
            col("parsed.confidence_score"),
            col("parsed.detected_text_summary"),
            col("parsed.llm_raw_result"),
            col("parsed.error_message")
        )
    )

    df_parsed_path = f"/Volumes/1_inland/sectra/vone/{str(dbutils.widgets.get('output_parquet'))}_parsed_batch{batch_idx}.parquet"

    df_parsed.write.mode("overwrite").parquet(df_parsed_path)

```
